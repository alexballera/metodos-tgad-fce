1. Una empresa de bebidas quiere estimar el volumen promedio real servido por una llenadora automática. Toma una muestra de 64 botellas y obtiene una media muestral $\bar{x} = 52{,}0$ ml. El desvío poblacional del proceso es conocido y vale $\sigma = 8{,}0$ ml. ¿Cuál es el intervalo de confianza del 95% para la media poblacional $\mu$?

**Datos de la muestra:**

$n = 64$

$\bar{x} = 52.0 \text{ ml}$

$\sigma = 8.0 \text{ ml}$

Nivel de confianza: $95\%$

$\alpha = 0.05$

Valor crítico de la distribución normal:

$z_{\alpha/2} = z_{0.025} = 1.96$

---

**Fórmula del intervalo de confianza**

Cuando el desvío poblacional es conocido, el intervalo de confianza para la media es:

$$IC = \bar{x} \pm z_{\alpha/2}\frac{\sigma}{\sqrt{n}}$$

---

**Desvío estándar**

$SE = \dfrac{\sigma}{\sqrt{n}} = \dfrac{8}{\sqrt{64}} = \dfrac{8}{8} = 1$

---

**Margen de error**

$ME = z_{\alpha/2} \cdot SE = 1.96 \times 1 = 1.96$

---

**Intervalo de confianza**

$IC = 52.0 \pm 1.96$

$IC = (50.04,\; 53.96)$

---

**Interpretación**

Con un nivel de confianza del 95%, se estima que el volumen promedio real servido por la llenadora automática se encuentra entre **50.04 ml y 53.96 ml**.

```python
# --- Datos de entrada ---
n = 64
media_muestral = 52.0
sigma = 8.0
confianza = 0.95

# --- Cálculo del Error Estándar ---
# Representa la desviación de la distribución de medias muestrales
error_estandar = sigma / np.sqrt(n)

# --- Cálculo del Intervalo con Scipy ---
# Usamos 'norm' porque el desvío poblacional (sigma) es conocido
# loc: media muestral | scale: error estándar
limite_inf, limite_sup = stats.norm.interval(confianza, loc=media_muestral, scale=error_estandar)

# --- Respuesta Final ---
print(f"Respuesta: El intervalo de confianza del {confianza*100}% para la media poblacional es [{limite_inf:.2f} ml - {limite_sup:.2f} ml].")
```

2. Un contact center quiere estimar el tiempo promedio de atención. Se mide una muestra de 100 llamadas: $\bar{x} = 120{,}0$ segundos. Se sabe que $\sigma = 15{,}0$ segundos. ¿Cuál es el intervalo de confianza del 90% para $\mu$?

**Datos del problema**

$n = 100$

$\bar{x} = 120.0 \text{ segundos}$

$\sigma = 15.0 \text{ segundos}$

Nivel de confianza: $90\%$

$\alpha = 0.10$, $\quad \alpha/2 = 0.05$

Valor crítico de la distribución normal:

$z_{\alpha/2} = z_{0.05} = 1.645$

---

**Fórmula del intervalo de confianza**

Cuando el desvío poblacional es conocido, el intervalo de confianza para la media poblacional se calcula como:

$$IC = \bar{x} \pm z_{\alpha/2}\frac{\sigma}{\sqrt{n}}$$

---

**Cálculo del error estándar**

$SE = \dfrac{\sigma}{\sqrt{n}} = \dfrac{15}{\sqrt{100}} = \dfrac{15}{10} = 1.5$

---

**Cálculo del margen de error**

$ME = z_{\alpha/2} \cdot SE = 1.645 \times 1.5 = 2.4675$

---

**Intervalo de confianza**

$IC = 120.0 \pm 2.4675$

$IC = (117.53,\; 122.47)$

---

**Interpretación**

Con un nivel de confianza del $90\%$, se estima que el tiempo promedio real de atención del contact center se encuentra entre **117.53 segundos y 122.47 segundos**.

```python
# --- Ingreso de parámetros del problema ---
media_muestral = 120.0
sigma = 15.0
n = 100
confianza = 0.90

# --- Cálculo del Error Estándar ---
# Es la desviación estándar de la distribución de la media
error_estandar = sigma / np.sqrt(n)

# --- Construcción del Intervalo ---
# Usamos stats.norm.interval porque conocemos el desvío poblacional (sigma)
# 'loc' es el centro (media) y 'scale' es la dispersión de la media (error estándar)
limite_inf, limite_sup = stats.norm.interval(confianza, loc=media_muestral, scale=error_estandar)

# --- Respuesta final ---
print(f"RESPUESTA: El intervalo de confianza del {confianza*100}% para el tiempo promedio de atención (μ) es [{limite_inf:.2f} s - {limite_sup:.2f} s].")
```

3. Una fintech está evaluando el tiempo de respuesta de su API. Su objetivo es que el tiempo promedio sea de 5 segundos. Para comprobarlo, tomaron 12 mediciones y obtuvieron un promedio de 5.4 segundos, con una desviación estándar de 0.9 segundos. Quieren determinar si hay evidencia suficiente para afirmar que el tiempo promedio es mayor a 5 segundos, asumiendo un nivel de significancia del 10%. ¿Cuál es el valor del estadístico de prueba calculado y qué decisión tomarían?

**Datos del problema**

$n = 12$, $\quad \bar{x} = 5.4 \text{ segundos}$, $\quad s = 0.9 \text{ segundos}$, $\quad \mu_0 = 5 \text{ segundos}$

Nivel de significancia: $\alpha = 0.10$, $\quad$ Grados de libertad: $gl = n-1 = 11$

---

**Planteo de hipótesis**

Hipótesis nula: $H_0 : \mu = 5$

Hipótesis alternativa (prueba unilateral derecha): $H_1 : \mu > 5$

---

**Estadístico de prueba**

Como el desvío poblacional es desconocido y el tamaño de muestra es pequeño, se utiliza la distribución t de Student:

$$t = \frac{\bar{x}-\mu_0}{s/\sqrt{n}}$$

---

**Cálculo**

$SE = \dfrac{0.9}{\sqrt{12}} = \dfrac{0.9}{3.464} \approx 0.2598$

$t = \dfrac{5.4-5}{0.2598} = \dfrac{0.4}{0.2598} \approx 1.54$

---

**Regla de decisión**

Para una prueba unilateral con $\alpha = 0.10$ y $gl = 11$: $t_{crit} \approx 1.363$

Si $t > t_{crit}$ entonces se rechaza $H_0$.

Como $1.54 > 1.363$, se **rechaza la hipótesis nula**.

---

**Conclusión**

Con un nivel de significancia del $10\%$, existe evidencia estadística suficiente para afirmar que el tiempo promedio de respuesta de la API es **mayor a 5 segundos**.

```python
# --- Definición de parámetros muestrales ---
n = 12
media_muestral = 5.4
media_hipotesis = 5.0
desvio_muestral = 0.9
alfa = 0.10
df = n - 1  # Grados de libertad para T de Student

# --- Cálculo del Estadístico de Prueba (t-obs) ---
error_estandar = desvio_muestral / np.sqrt(n)
t_obs = (media_muestral - media_hipotesis) / error_estandar

# --- Obtención de Valor Crítico y P-valor ---
t_critico = stats.t.ppf(1 - alfa, df)  # Valor de corte para el 10%
p_valor = 1 - stats.t.cdf(t_obs, df)  # Probabilidad de obtener t_obs o mayor bajo H0

# --- Decisión lógica ---
decision = "Rechazar H0" if t_obs > t_critico else "No rechazar H0"

# --- Respuesta final ---
print(f"RESULTADO: t-observado = {t_obs:.3f} | t-crítico = {t_critico:.3f} | p-valor = {p_valor:.4f}")
print(f"DECISIÓN: Con un nivel de significancia del {alfa*100}%, la decisión es: {decision}.")
```

7. Una empresa de telecomunicaciones quiere conocer cuánto gastan en promedio sus clientes por mes. El objetivo es que el error máximo en la estimación sea de solo 2 unidades monetarias. Basándose en datos históricos, saben que la desviación estándar del gasto es de 12 unidades. Exigen un nivel de confianza del 95%. ¿Cuál es la cantidad mínima de clientes que deben incluir en su muestra?

**Datos del problema**

Error máximo permitido: $E = 2$

Desvío estándar poblacional: $\sigma = 12$

Nivel de confianza: $95\%$ $\Rightarrow$ $\alpha = 0.05$ $\Rightarrow$ $z_{\alpha/2} = 1.96$

---

**Fórmula para el tamaño de muestra**

Cuando se desea estimar una media con un error máximo $E$ y se conoce el desvío poblacional:

$$n = \left(\frac{z_{\alpha/2}\,\sigma}{E}\right)^2$$

---

**Cálculo**

$$n = \left(\frac{1.96 \times 12}{2}\right)^2 = (11.76)^2 = 138.30 \Rightarrow n = 139$$

---

**Interpretación**

La empresa debe encuestar **al menos 139 clientes** para estimar el gasto promedio mensual con un error máximo de **2 unidades monetarias** y un **95% de confianza**.

```python
# --- Definición de parámetros del problema ---
sigma = 12.0
error_maximo = 2.0
confianza = 0.95

# --- Cálculo del Valor Crítico Z ---
# Calculamos Z para el 95% (bilateral)
z = stats.norm.ppf(1 - (1 - confianza) / 2)

# --- Cálculo del Tamaño de Muestra (n) ---
n_exacto = (z * sigma / error_maximo) ** 2
n_final = int(np.ceil(n_exacto))  # Redondeo hacia arriba

# --- Respuesta final ---
print(f"RESULTADO: La muestra mínima requerida es de {n_final} clientes.")
print(f"Detalles técnicos: Z = {z:.3f}, n calculado = {n_exacto:.4f}")
```

8. Una cadena de tiendas (retail) quiere estimar cuánto gasta en promedio cada cliente por compra (ticket promedio). La variabilidad histórica del gasto (desvío estándar) es de 9 unidades monetarias. Exigen un nivel de confianza del 90% y necesitan que el intervalo total tenga una amplitud de 6 unidades (margen de error de ±3 unidades). ¿Cuál es la cantidad mínima de clientes que deben incluir en su muestra?

**Datos del problema**

$\sigma = 9$, $\quad$ Nivel de confianza: $90\%$ $\Rightarrow$ $z_{\alpha/2} = 1.645$

Amplitud total deseada: $A = 6$ $\Rightarrow$ Margen de error: $E = A/2 = 3$

---

**Fórmula para el tamaño de muestra**

$$n = \left(\frac{z_{\alpha/2}\,\sigma}{E}\right)^2$$

---

**Cálculo**

$$n = \left(\frac{1.645 \times 9}{3}\right)^2 = (4.935)^2 = 24.35 \Rightarrow n = 25$$

---

**Interpretación**

La cadena de tiendas debe observar **al menos 25 clientes** para estimar el ticket promedio con un **90% de confianza** y con un intervalo cuya amplitud total sea de **6 unidades monetarias** (es decir, $\pm 3$ alrededor de la media estimada).

```python
# --- Definición de parámetros del requisito ---
sigma = 9.0
confianza = 0.90
amplitud_deseada = 6.0
margen_error = amplitud_deseada / 2

# --- Cálculo del Valor Crítico Z ---
# Obtenemos el valor Z para el 90% de confianza
z = stats.norm.ppf(1 - (1 - confianza) / 2)

# --- Cálculo del Tamaño de Muestra (n) ---
n_exacto = (z * sigma / margen_error) ** 2
n_final = int(np.ceil(n_exacto))  # Redondeo hacia arriba al entero superior

# --- Respuesta final ---
print(f"RESULTADO: El tamaño de muestra necesario es de {n_final} clientes.")
print(f"Cálculo técnico: n exacto = {n_exacto:.4f}, Z = {z:.3f}")
```

9. Una empresa manufacturera produce piezas metálicas y necesita controlar la variabilidad dimensional de su proceso. Para evaluar este control de calidad, el equipo de ingeniería tomó una muestra aleatoria de 20 piezas de la línea de producción. Tras las mediciones, calcularon una varianza muestral ($s^2$) de 25,0. La gerencia técnica necesita conocer, con un 95% de confianza, entre qué valores oscila la verdadera varianza poblacional ($\sigma^2$) del proceso para decidir si deben calibrar las máquinas.

**Datos del problema**

$n = 20$, $\quad s^2 = 25.0$, $\quad$ Nivel de confianza: $95\%$ $\Rightarrow$ $\alpha = 0.05$

Grados de libertad: $gl = n - 1 = 19$

---

**Fórmula del intervalo de confianza para la varianza**

Cuando se desea estimar la varianza poblacional, el intervalo se construye con la distribución chi-cuadrado:

$$IC_{\sigma^2} = \left(\frac{(n-1)s^2}{\chi^2_{\alpha/2,\;gl}},\; \frac{(n-1)s^2}{\chi^2_{1-\alpha/2,\;gl}}\right)$$

---

**Valores críticos** para $gl = 19$ y $\alpha = 0.05$:

$\chi^2_{0.025,19} \approx 32.852$, $\quad \chi^2_{0.975,19} \approx 8.907$

$(n-1)s^2 = 19 \times 25 = 475$

Límite inferior: $\dfrac{475}{32.852} \approx 14.46$

Límite superior: $\dfrac{475}{8.907} \approx 53.33$

---

**Intervalo de confianza**

$IC_{\sigma^2} = (14.46,\; 53.33)$

---

**Interpretación**

Con un nivel de confianza del $95\%$, la varianza poblacional del espesor de las piezas se encuentra entre **14.46 y 53.33**. Esto permite evaluar si la variabilidad del proceso es aceptable o si es necesario recalibrar las máquinas.

```python
# --- Definición de parámetros del proceso ---
n = 20
varianza_muestral = 25.0
confianza = 0.95
df = n - 1  # Grados de libertad

# --- Obtención de valores críticos Chi-cuadrado ---
# Buscamos los puntos de corte para dejar 2.5% en cada cola
chi2_inf = stats.chi2.ppf((1 - confianza) / 2, df)  # Cola izquierda
chi2_sup = stats.chi2.ppf((1 + confianza) / 2, df)  # Cola derecha

# --- Cálculo de los límites del intervalo ---
# La fórmula invierte los valores de chi2 para los límites inf y sup
limite_inf = (df * varianza_muestral) / chi2_sup
limite_sup = (df * varianza_muestral) / chi2_inf

# --- Respuesta final ---
print(f"ANÁLISIS DE VARIABILIDAD:")
print(f"El intervalo de confianza del {confianza*100}% para la varianza poblacional (sigma^2)")
print(f"es de {limite_inf:.2f} a {limite_sup:.2f}.")
```

13. El equipo de Marketing de una tienda online está analizando la efectividad de sus campañas de retención. Seleccionaron una muestra aleatoria de 400 clientes que realizaron su primera compra hace exactamente dos meses. Al revisar el historial de transacciones, detectaron que 92 de esos clientes volvieron a comprar dentro de los primeros 30 días. La gerencia solicita un intervalo de confianza del 95% para la verdadera proporción de clientes que recompran en toda la base de datos.

**Datos del problema**

$n = 400$, $\quad x = 92$

Proporción muestral: $\hat{p} = \dfrac{92}{400} = 0.23$

Nivel de confianza: $95\%$ $\Rightarrow$ $z_{\alpha/2} = 1.96$

---

**Fórmula del intervalo de confianza para una proporción**

$$IC = \hat{p} \pm z_{\alpha/2}\sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$$

---

**Cálculo del error estándar**

$SE = \sqrt{\dfrac{0.23 \times 0.77}{400}} = \sqrt{0.00044275} \approx 0.02105$

**Margen de error:** $ME = 1.96 \times 0.02105 \approx 0.0413$

**Intervalo de confianza:** $IC = 0.23 \pm 0.0413 = (0.1887,\; 0.2713)$

---

**Interpretación**

Con un nivel de confianza del $95\%$, la proporción real de clientes que recompran se encuentra entre **18.9\% y 27.1\%**.

```python
# --- Definición de parámetros muestrales ---
n = 400
exitos = 92
confianza = 0.95

# --- Cálculo del Intervalo para Proporciones (Método Normal) ---
# Usamos alpha = 1 - confianza (0.05 para un 95%)
limite_inf, limite_sup = proportion_confint(count=exitos,
                                            nobs=n,
                                            alpha=(1 - confianza),
                                            method='normal')

# --- Respuesta final ---
print(f"ANÁLISIS DE FIDELIZACIÓN:")
print(f"En una muestra de {n} clientes, la tasa de recompra es del {(exitos/n):.1%}.")
print(f"El intervalo de confianza del {confianza*100}% para la proporción real")
print(f"es de {limite_inf:.4f} a {limite_sup:.4f} (aprox. {limite_inf:.1%} - {limite_sup:.1%}).")
```

17. Un banco digital está diseñando el lanzamiento de una nueva tarjeta de crédito 100% virtual. Como es un producto totalmente innovador, el banco no tiene datos históricos. Para no subestimar el esfuerzo necesario, deciden adoptar la postura más conservadora ($p = 0{,}50$). Exigen un nivel de confianza del 99% y un margen de error de apenas el 3% ($0{,}03$). ¿Cuántos usuarios deben encuestar como mínimo?

**Datos del problema**

Proporción asumida (caso más conservador): $p = 0.50$

Margen de error: $E = 0.03$

Nivel de confianza: $99\%$ $\Rightarrow$ $z_{\alpha/2} = z_{0.005} = 2.576$

---

**Fórmula para el tamaño de muestra (proporciones)**

Cuando no se conoce la proporción poblacional, se utiliza el caso más conservador $p=0.5$:

$$n = \frac{z_{\alpha/2}^2 \cdot p(1-p)}{E^2}$$

---

**Cálculo**

$$n = \frac{(2.576)^2 \times 0.5 \times 0.5}{(0.03)^2} = \frac{6.635 \times 0.25}{0.0009} = \frac{1.65875}{0.0009} = 1843.06 \Rightarrow n = 1844$$

---

**Interpretación**

El banco debe encuestar **al menos 1844 usuarios** para estimar la proporción de adopción de la nueva tarjeta con un **99% de confianza** y un margen de error de **±3\%**, bajo el escenario más conservador.

```python
# --- Definición de parámetros (Caso Conservador) ---
p_conservador = 0.50
q_conservador = 1 - p_conservador
error_maximo = 0.03
confianza = 0.99

# --- Cálculo del Valor Crítico Z ---
# Para 99% de confianza, z es aproximadamente 2.576
z = stats.norm.ppf(1 - (1 - confianza) / 2)

# --- Cálculo del Tamaño de Muestra (n) ---
n_exacto = (z ** 2 * p_conservador * q_conservador) / (error_maximo ** 2)
n_final = int(np.ceil(n_exacto))  # Redondeo al entero superior

# --- Respuesta final ---
print(f"ESTUDIO DE BANCA DIGITAL:")
print(f"Bajo el caso más conservador (p=0.5), con confianza del {confianza*100}%")
print(f"y un error del {error_maximo*100}%, la muestra mínima es: {n_final} usuarios.")
print(f"Valor crítico Z utilizado: {z:.3f}")
```

20. El equipo de Growth Marketing ha lanzado dos versiones de una página de aterrizaje para una nueva campaña de suscripción. La Página A utilizó un diseño con colores cálidos y testimonios, mientras que la Página B es minimalista y directa. Tras una semana de tráfico, los datos muestran que la Página A tuvo 78 conversiones sobre 300 visitas, mientras que la Página B logró 50 sobre 250. Marketing necesita saber, con un 95% de confianza, cuál es el rango de la diferencia de rendimiento entre ambas ($\pi_A - \pi_B$) para decidir cuál dejar como versión definitiva.

**Datos observados:**

Página A: $n_A = 300$, $x_A = 78$ $\Rightarrow \hat{p}_A = 78/300 = 0.26$

Página B: $n_B = 250$, $x_B = 50$ $\Rightarrow \hat{p}_B = 50/250 = 0.20$

Diferencia observada: $\hat{p}_A - \hat{p}_B = 0.26 - 0.20 = 0.06$

Nivel de confianza: $95\%$ $\Rightarrow$ $z_{\alpha/2} = 1.96$

---

### Error estándar de la diferencia de proporciones

$$SE = \sqrt{\frac{\hat{p}_A(1-\hat{p}_A)}{n_A} + \frac{\hat{p}_B(1-\hat{p}_B)}{n_B}} = \sqrt{0.000641 + 0.000640} \approx 0.0358$$

### Margen de error

$ME = 1.96 \times 0.0358 \approx 0.070$

### Intervalo de confianza

$IC_{95\%} = 0.06 \pm 0.070 = (-0.010,\; 0.130)$

---

### Interpretación

Con un nivel de confianza del 95%, la diferencia real entre las tasas de conversión de las páginas A y B se encuentra aproximadamente entre $-1.0\%$ y $13.0\%$.

Como el intervalo incluye el valor 0, no se puede afirmar con un 95% de confianza que una página tenga una tasa de conversión mayor que la otra.

```python
# --- Datos Pagina A ---
n = 300
exito_a = 78
p_a = 0.26

# --- Datos Pagina B ---
m = 250
exito_b = 50
p_b = 0.20

confianza = 0.95
df = n + m - 2

# --- Cálculo del Error Estándar de la Diferencia de Proporciones ---
s2_p = np.sqrt(p_a * (1 - p_a) / n + p_b * (1 - p_b) / m)

# --- Valor Crítico T ---
t_crit = stats.t.ppf(1 - (1 - confianza) / 2, df)
margen_error = s2_p * t_crit

# --- Intervalo de Confianza ---
dif_muestral = p_a - p_b
lim_inf, lim_sup = dif_muestral - margen_error, dif_muestral + margen_error

print(f"ANÁLISIS PERFORMANCE DE WEB:")
print(f"Diferencia de proporciones: {dif_muestral}")
print(f"Error estándar (s2_p): {s2_p:.4f}")
print(f"Intervalo 95% T: [{lim_inf:.2f} , {lim_sup:.2f}]")
```
